```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
from lloyd.lloyd import Field
import scipy
```

```python
from scipy.stats import qmc
sampler = qmc.Halton(d=2, scramble=True)
points = sampler.random(n=10)

vor = Voronoi(points)

fig = voronoi_plot_2d(vor)
plt.show
```

```python
from lloyd import Field
from scipy.spatial import voronoi_plot_2d

# build a model we can use to run lloyd relaxation on the data points
field = Field(points)

# plot the initial voronoi map
voronoi_plot_2d(field.voronoi, show_vertices=False, line_colors='black', line_alpha=0.6, point_size=2)

# run lloyd relaxation several times until desired criteria are met
while scipy.spatial.distance.pdist(field.get_points()).min() < 0.17:

  # the .relax() method performs lloyd relaxation, which spaces the points apart
  field.relax()
  
  #plot the updated points and voronoi map
  p = voronoi_plot_2d(field.voronoi, show_vertices=False, line_colors='black', line_alpha=0.6, point_size=2)
```
